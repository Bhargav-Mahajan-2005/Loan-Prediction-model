```python
def create_dummies (df,column_name):
  dummies =pd.get_dummies(df[column_name],prefix=column_name)
  df=pd.concat([df,dummies],axis=1 )     #concat means all extra columns due to dummies concat with  original data frame
  return df
```

```python
for column in ['Education','Gender','Married']:
  df=create_dummies(df,column)
```

```python
df.drop(['Loan_ID','Gender','Married','Education','Property_Area','Loan_Status','Self_Employed','Dependents'],axis=1,inplace=True)
```

```python
X=df.drop('Loan_Amount_Term',axis=1)
Y=df['Loan_Amount_Term']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import RFE
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.linear_model import LinearRegression

x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
LR.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred=LR.predict(x_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
from sklearn.svm import SVC   #support vector classifier
SVC=SVC()    #parameters(Hyper parameters)-Tuning
SVC.fit(x_train,y_train)
```

```python
y_pred=SVC.predict(x_test)
```

```python
accuracy_score(y_pred,y_test)
```
